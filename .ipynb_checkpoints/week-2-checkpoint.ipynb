## 2. Convert json files into Pandas.Dataframe

```python
Json1 = [{'name': 'JOHN', 'age': 27}, {'name': 'HENRY2', 'age': 14}]

Json2 = { 1:{'name': 'vikash', 'age': 27}, 2:{'name': 'Satyam', 'age': 14} }

data = [{'id': 1, 'name': "Cole Volk", 'fitness': {'height': 130, 'weight': 60}}, 
        {'name': "Mose Reg", 'fitness': {'height': 130, 'weight': 60}}, 
        {'id': 2, 'name': 'Faye Raker', 'fitness': {'height': 130, 'weight': 60}}]
```

```python
# Create df1 from Json1
df1=pd.DataFrame(Json1)
df1
```

```python
# Create df2 from Json2
df2=pd.DataFrame(Json2).T
df2
```

```python
# Concatenate df1 and df2 data frame
pd.concat([df1, df2], sort=False)
```

```python
# Create df dataframe from data
df=pd.DataFrame(data)
df
```

```python
# Create df4 Dataframe from df.fitness
df4=pd.DataFrame([i for i in df.fitness])
df4
```

```python
# Concatenate df and df4 data frame
pd.concat([df,df4],axis=1).drop('fitness',axis=1)
```

## 3. Draw the graphs below for the given data:

```python
# import libraries
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
from pylab import rcParams
```

```python
labels = ['UK', 'USA', 'GER', 'JAP', 'FRA']
age_men = [40, 35, 45, 53, 28]
age_women = [35, 32, 48, 48, 31]
```

```python
# Draw Age means of mens with matplotlib
# tips: https://matplotlib.org/stable/gallery/lines_bars_and_markers/barchart.html#sphx-glr-gallery-lines-bars-and-markers-barchart-py

x = np.arange(len(labels))  # the label locations
width = 0.35  # the width of the bars

fig, ax = plt.subplots()
rects1 = ax.bar(x, age_men, width, label='Men')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Ages')
ax.set_title('Age Means by Gender')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()

def autolabel(rects):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')


autolabel(rects1)
fig.tight_layout()
plt.show()
```

```python
# Draw Age means of womens with matplotlib
fig, ax = plt.subplots()
rects2 = ax.bar(x, age_women, width, label='Women')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Ages')
ax.set_title('Age Means by Gender')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()

autolabel(rects2)
fig.tight_layout()
plt.show()
```

```python
# Draw Age means by Gender with matplotlib
fig, ax = plt.subplots()
rects1 = ax.bar(x- width/2, age_men, width, label='Men')
rects2 = ax.bar(x+ width/2, age_women, width, label='Women')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Ages')
ax.set_title('Age Means by Gender')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()

def autolabel(rects):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')


autolabel(rects1)
autolabel(rects2)
fig.tight_layout()
plt.show()
```

```python
# Alternative Solution with seaborn
import seaborn as sns

labels = ['UK', 'USA', 'GER', 'JAP', 'FRA']
age_men = [40, 35, 45, 53, 28]
age_women = [35, 32, 48, 48, 31]
```

```python
df_m = pd.DataFrame({'Country':labels,'Ages':age_men,'Gender':['Male']*len(age_men)})
df_f = pd.DataFrame({'Country':labels,'Ages':age_women,'Gender':['Female']*len(age_women)})
```

```python
df= pd.concat([df_m, df_f]).reset_index().drop('index',axis=1)
df
```

```python
rcParams['figure.figsize'] = 6,15 
def labels(ax):
    for bar in ax.patches: 
        ax.annotate(format(bar.get_height(), '.0f'),  
                       (bar.get_x() + bar.get_width() / 2,  
                        bar.get_height()), ha='center', va='center', 
                       size=10, xytext=(0, 8), 
                       textcoords='offset points') 
        
plt.subplot(311)
ax = sns.barplot(x='Country',y='Ages',data= df[df.Gender=="Male"],hue='Gender')
labels(ax)
plt.xlabel("Country", size=10) 
plt.ylabel("Ages", size=10) 
plt.title("Age Means by Gender") 
plt.show()

plt.subplot(312)
ax = sns.barplot(x='Country',y='Ages',data= df[df.Gender=="Female"],hue='Gender')
labels(ax)
plt.xlabel("Country", size=10)  
plt.ylabel("Ages", size=10) 
plt.title("Age Means by Gender") 
plt.show()

plt.subplot(313)
ax = sns.barplot(x='Country',y='Ages',data= df,hue='Gender')
labels(ax)
plt.xlabel("Country", size=10)  
plt.ylabel("Ages", size=10) 
plt.title("Age Means by Gender") 
plt.show()
```
